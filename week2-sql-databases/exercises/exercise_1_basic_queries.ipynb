````python
# Write your query here
query_5 = """

"""
run_query(query_5)
````

> **HINT** - scroll down only if stuck  
> Use `EXTRACT(YEAR FROM hire_date)` to get the year from a date column. Then use `COUNT(*)` with `GROUP BY` on the extracted year. Alias the extracted year as `hire_year` for readability.

<details>
<summary><strong>SOLUTION - Problem 5</strong> (click to reveal)</summary>

```sql
SELECT EXTRACT(YEAR FROM hire_date) AS hire_year,
       COUNT(*) AS employee_count
FROM company.employees
GROUP BY EXTRACT(YEAR FROM hire_date)
ORDER BY hire_year;
```

**Explanation:** `EXTRACT(YEAR FROM hire_date)` pulls the year from each hire date. We group by this extracted value and count the rows in each group. The `GROUP BY` must match the expression in the `SELECT` (or we can use the alias in some databases, but PostgreSQL requires the full expression or the alias defined in `SELECT`). Ordering by `hire_year` shows the hiring trend chronologically.
</details>

## Problem 6: Vowel-Starting First Names

**Business Question:**  
For a fun company directory exercise, find all employees whose first name starts with a vowel (A, E, I, O, or U, case-insensitive). Show their full name as a single column (first and last name combined with a space in between), their department ID, and hire date. Order by hire date from most recent to oldest.

````python
# Write your query here
query_6 = """

"""
run_query(query_6)
````

> **HINT** - scroll down only if stuck  
> Use `CONCAT(first_name, ' ', last_name)` to combine names. For the vowel check, you can use `LEFT(LOWER(first_name), 1) IN ('a', 'e', 'i', 'o', 'u')` or `SUBSTRING(first_name, 1, 1)` with `ILIKE`.

<details>
<summary><strong>SOLUTION - Problem 6</strong> (click to reveal)</summary>

```sql
SELECT CONCAT(first_name, ' ', last_name) AS full_name,
       department_id,
       hire_date
FROM company.employees
WHERE LOWER(LEFT(first_name, 1)) IN ('a', 'e', 'i', 'o', 'u')
ORDER BY hire_date DESC;
```

**Explanation:** `CONCAT(first_name, ' ', last_name)` joins the first and last name with a space. `LEFT(first_name, 1)` extracts the first character, and `LOWER()` ensures the check is case-insensitive. The `IN` operator then checks if that character is one of the five vowels. Finally, `ORDER BY hire_date DESC` shows the most recently hired employees first.

**Alternative approach:** You could also write the WHERE clause as:
```sql
WHERE first_name ILIKE 'A%' 
   OR first_name ILIKE 'E%' 
   OR first_name ILIKE 'I%' 
   OR first_name ILIKE 'O%' 
   OR first_name ILIKE 'U%'
```
`ILIKE` is PostgreSQL-specific and performs case-insensitive matching.
</details>
